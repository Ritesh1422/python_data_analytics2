```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
t=np.linspace(0,3*np.pi,50)
x=np.sin(t)
y=np.cos(t)
```

```python
plt.figure(figsize=(15,5))
plt.plot(t,x, 'ro-')
plt.plot(t,y, 'h-')
```
