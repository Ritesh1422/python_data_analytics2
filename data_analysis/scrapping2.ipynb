```python
from dputils.scrape import Tag, Scraper
```

```python
url="https://timesofindia.indiatimes.com/news"
```

```python
news=Scraper(url)
```

```python
target=Tag(cls="listing4 clearfix")
item=Tag("li")
title=Tag('span', cls="w_tle")
desc=Tag('span',cls="w_desc")
```

```python
news.get_all(target,item,title=title,description=desc)
```

```python
result=news.get_all(target,item,title=title,description=desc)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(result)
```
