binning data into groups 
. monetary >10000=>high
. 10000>monetary>5000=>medium
. 5000>monetary>1000=>low
. monetary<1000=>very low


```python
def categorize_monetary_value(amt):
    if amt>10000:
        return 'high'
    elif amt>5000:
        return 'medium'
    elif amt>1000:
        return'low'
    else:
        'very low'

categorize_monetary_value(5500)        
```
