```python
import pandas as pd
```

```python
host='localhost'
port=3306
username='root'
password='root'
database='sales'
protocol='mysql://'
conn=f'{protocol}{username}:{password}@{host}:{port}/{database}'
print(conn)
```

```python
querry='''select
    count(cid)as total,
    state,
    if(count(cid)>=10,"👌","X")as importance
FROM customers 
group by state
order by importance;
   '''

df_result=pd.read_sql(querry,con=conn)
df_result.total.plot(style='o-',figsize=(10,5))
```
